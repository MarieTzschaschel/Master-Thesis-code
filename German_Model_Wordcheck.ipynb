## Examination of word representations for the German model

In this project we examine whether all relevant nouns are represented by a word vector with the German Word2vec model. This model is completely pretrained, therefore no adaptations can be made. Further information about the model is available here: https://devmount.github.io/GermanWordEmbeddings/

We start by downloading the relevant libraries, the pretrained model and the experimental words.

```python
import pandas as pd
import gensim
import matplotlib.pyplot as plt
```

```python
trained_model = gensim.models.KeyedVectors.load_word2vec_format('german.model', binary=True)
```

```python
df = pd.read_csv('vasishth_2010_Ger.csv', sep=';')
```

```python
df
```

We store all the words that have an associated word vector into the variable word_dict.

```python
word_dict = trained_model.wv.vocab
```

Next, we create a function that iterates over the experimental words and checks if each word is present in word_dict. Or in other words, if each word is represented by a word vector in our model.

```python
df['w1de_inW2V'] = df.w1de.apply(lambda x: x in word_dict)
df['w2de_inW2V'] = df.w2de.apply(lambda x: x in word_dict)
df['w3de_inW2V'] = df.w3de.apply(lambda x: x in word_dict)
df['w2de_i_inW2V'] = df.w2de_i.apply(lambda x: x in word_dict)
```

```python
df
```

We can see that all the relevant words are represented in our word2vec model.
